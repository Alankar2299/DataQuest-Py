### define and use price_label function

```python
#define function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df = ords_prods_merged
```

## Apply the function to ords_prods_merged to create a new dataframe Price_range

```python
df['price_range']= df.apply(price_label,axis=1)
```

## Further divide the busiest day column in more categories and rename it as 'busiest days'

```python
# calculate busiest days and create column
result_busy = []
for vl in df["orders_day_of_week"] :
    if vl == 0 :
        result_busy.append("Number 1 busiest day")
    elif vl == 1 :
        result_busy.append("Number 2 busiest day") 
    elif (vl == 6 ) or (vl == 2 ) or (vl == 5 ):
        result_busy.append("Regularly busy") 
    elif vl == 3 :
        result_busy.append("Number 2 slowest day")  
    else :
        result_busy.append("Number 1 slowest day")
```

```python
df['Busiest_days']=result_busy
```

## further division based on number of orders

The following criteria will be used for defining the time periods

Most orders = hours with more than 12,000 order count 
Average orders = hours between 9000 to 11999 order count  
Fewest orders = hours below 8999 order count 

```python
df['Order_time_of_day'].value_counts(dropna = False)
```

```python
hr = []

for vl in df["Order_time_of_day"] :
    if vl in ([14,11,12,13,10,15,16,9,17,8,18]) :
        hr.append("Most orders")
    elif vl in ([19,20]) :
        hr.append("Average orders") 
    else :
        hr.append("Fewest orders")
```

```python
hr
```

```python
df['busiest_period_of_day']= hr
```
